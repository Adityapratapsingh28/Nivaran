```python
X = standardized_data
Y = df['copd']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
X.shape, X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Initialize the RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100,   # Number of trees in the forest
    max_depth=None,     # Maximum depth of the trees
    random_state=42,    # Seed for reproducibility
    criterion='gini')

# Train the model
rf.fit(X_train, Y_train)

# Make predictions
y_pred = rf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(Y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
import numpy as np

def get_user_input():
    features = [
        'AGE', 'PackHistory', 'MWT1Best', 'FEV1', 'FVC', 'CAT', 'HAD', 'SGRQ',
        'gender', 'smoking', 'Diabetes', 'muscular', 'hypertension',
        'AtrialFib', 'IHD'
    ]
    
    # Example input values; replace with actual user input
    input_values = [
    65,      # AGE
    50,      # PackHistory
    50,      # MWT1Best
    30,      # FEV1
    60,      # FVC
    30,      # CAT
    15,      # HAD
    40,      # SGRQ
    1,       # gender
    1,       # smoking
    1,       # Diabetes
    1,       # muscular
    1,       # hypertension
    1,       # AtrialFib
    1        # IHD
]
    
    return input_values

# Get user input
input_data = get_user_input()

# Reshape the input data for the model
input_data_reshaped = np.array(input_data).reshape(1, -1)

# Make prediction
prediction = rf.predict(input_data_reshaped)
prediction_proba = rf.predict_proba(input_data_reshaped)

# Interpret result
gold_stages = {
    1: "GOLD 1: Mild COPD",
    2: "GOLD 2: Moderate COPD",
    3: "GOLD 3: Severe COPD",
    4: "GOLD 4: Very severe COPD"
}

predicted_class = prediction[0]
confidence_percentage = np.max(prediction_proba) * 100

if predicted_class in gold_stages:
    print(f"The model predicts: {gold_stages[predicted_class]}")
    print(f"Confidence: {confidence_percentage:.2f}%")
else:
    print("Invalid input")
```
